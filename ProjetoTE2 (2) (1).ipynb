### Processando os dados

Eliminação de algumas features

```python
df.drop(['Nº CV SIAFI/ SICONV',
         ' Nº CV Interno', 'Instituição Convenente',
         'Título do Projeto',
         'Cidade da Instituição Convenente',
         'Data Assinatura',
         'Data de Publicação no DOU',
         'Data de Fim da Vigência',
         'Ação, Edital ou Termo de Referência de Origem do Projeto'],
        axis=1, inplace=True)
```

```python
# colunas que contém valores nulos
sns.heatmap(df.isnull())
```

```python
# limpar os valores nulos
df.dropna(inplace=True)
sns.heatmap(df.isnull())
```

```python
def convert_float(data):
    data = data.replace('.', '').replace(',', '.')
    return np.float64(data)
```

```python
df["Valor Total do Projeto (R$)"] = df["Valor Total do Projeto (R$)"].apply(lambda data: convert_float(data))
df["Recursos de CP (R$)"] = df["Recursos de CP (R$)"].apply(lambda data: convert_float(data))
df["Recursos do FNMA (R$)"] = df["Recursos do FNMA (R$)"].apply(lambda data: convert_float(data))
```

```python
df
```

```python
# Aqui foi feita conversão dos valores Object para o tipo float

O objetivo é receber uma string que representa um número, ajustar o formato substituindo vírgulas por pontos e remover pontos extras, e então converter essa string para um número de ponto flutuante  (np.float64).
```

```python
from sklearn.preprocessing import LabelEncoder

# Crie uma instância do LabelEncoder
encoder = LabelEncoder()

df['Tema'] = encoder.fit_transform(df['Tema'])
df['UF'] = encoder.fit_transform(df['UF'])
df['Região Geográfica'] = encoder.fit_transform(df['Região Geográfica'])
df['Bioma'] = encoder.fit_transform(df['Bioma'])
df['Esfera Institucional'] = encoder.fit_transform(df['Esfera Institucional'])
df['Tipo de Seleção do Projeto'] = encoder.fit_transform(df['Tipo de Seleção do Projeto'])
```

```python
df
```

```python
#  Isso significa que ele ajusta o encoder aos valores únicos presentes na coluna 'Tema' e, em seguida, transforma esses valores em códigos numéricos. Essa transformação é atribuída de volta à coluna 'Tema' no DataFrame.

O mesmo processo é repetido para as outras colunas categóricas: 
    'UF', 'Região Geográfica', 'Bioma', 'Esfera Institucional', e 'Tipo de Seleção do Projeto'. 
Cada uma delas é ajustada e transformada separadamente.

A principal razão para usar o LabelEncoder é preparar os dados para algoritmos de machine learning que exigem entradas numéricas.
```

```python
<b>Testando o modelo</b>

Dimensionamento usado para ter uma escala comum. A partir os dados são divididos para treino e teste
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler

# Dividir os dados em conjuntos de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(df[['Ano', 'Tema', 'UF', 'Região Geográfica', 'Bioma',
       'Esfera Institucional', 'Recursos do FNMA (R$)', 'Recursos de CP (R$)', 'Tipo de Seleção do Projeto']], df[['Valor Total do Projeto (R$)']], test_size=0.2, random_state=42)

# Inicialize o StandardScaler
scaler = StandardScaler()

# Ajuste o scaler aos dados de treinamento e transforme-os
X_train_scaled = scaler.fit_transform(X_train)

# Transforme os dados de teste usando o mesmo scaler
X_test_scaled = scaler.transform(X_test)
```

```python
#Esse modelo é um exemplo de como usar a biblioteca scikit-learn para treinar um modelo de regressão linear em Python no dataset

'train_test_split' = Divide o conjunto de dados em conjuntos de treinamento e teste. 
Os conjuntos de recursos (X) incluem várias colunas do DataFrame df, 
e o conjunto de destino (y) é a coluna 'Valor Total do Projeto (R$)'.
```

### Regressão Linear

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Inicialize o modelo de regressão linear
modelo = LinearRegression()

# Treine o modelo com os dados de treinamento normalizados
modelo.fit(X_train_scaled, y_train)

# Faça previsões com o conjunto de teste normalizado
y_pred = modelo.predict(X_test_scaled)

# Calcula o Erro Quadrático Médio (MSE)
mse_linear = mean_squared_error(y_test, y_pred)

# Calcula o Erro Absoluto Médio (MAE)
mae_linear = mean_absolute_error(y_test, y_pred)

# Calcula o Coeficiente de Determinação (R²)
r2_linear = r2_score(y_test, y_pred)

# Exibe os resultados
print("Erro Quadrático Médio (MSE):", mse_linear)
print("Erro Absoluto Médio (MAE):", mae_linear)
print("Coeficiente de Determinação (R²):", r2_linear)
```

```python
#Os valores impressos indicam que o modelo tem um desempenho muito bom, com um erro quadrático médio baixo (143859007.29),
um erro absoluto médio relativamente baixo (700.70) e um coeficiente de determinação muito alto (0.9986), 
sugerindo que o modelo explica a variabilidade nos dados de teste muito bem.
```

### Árvore de Decisão de Regressão

```python
from sklearn.tree import DecisionTreeRegressor

# Inicializando o modelo de Regressão de Árvore de Decisão
reg_tree = DecisionTreeRegressor()

# Treinando o modelo
reg_tree.fit(X_train_scaled, y_train)

# Prevendo os valores no conjunto de teste
y_pred = reg_tree.predict(X_test)

# Calcula o Erro Quadrático Médio (MSE)
mse_tree = mean_squared_error(y_test, y_pred)

# Calcula o Erro Absoluto Médio (MAE)
mae_tree = mean_absolute_error(y_test, y_pred)

# Calcula o Coeficiente de Determinação (R²)
r2_tree = r2_score(y_test, y_pred)

mse = mean_squared_error(y_test, y_pred)
print("Erro Quadrático Médio (MSE):", mse_tree)
print("Erro Absoluto Médio (MAE):", mae_tree)
print("Coeficiente de Determinação (R²):", r2_tree)
```

### SVM Regressão

```python
from sklearn.svm import SVR

# Inicializando o modelo SVR
svm_regressor = SVR(kernel='linear')  # Escolha do kernel, pode ser 'linear', 'poly', 'rbf', etc.

# Treinando o modelo
svm_regressor.fit(X_train_scaled, y_train)

# Prevendo os valores no conjunto de teste
y_pred = svm_regressor.predict(X_test)

# Calcula o Erro Quadrático Médio (MSE)
mse_svm = mean_squared_error(y_test, y_pred)

# Calcula o Erro Absoluto Médio (MAE)
mae_svm = mean_absolute_error(y_test, y_pred)

# Calcula o Coeficiente de Determinação (R²)
r2_svm = r2_score(y_test, y_pred)

print("Erro Quadrático Médio (MSE):", mse_svm)
print("Erro Absoluto Médio (MAE):", mae_svm)
print("Coeficiente de Determinação (R²):", r2_svm)
```

```python
# Visualizando os resultados
labels = ['MSE']
values_tree = [mse_tree]
values_svm = [mse_svm]
values_linear = [mse_linear]  # Certifique-se de definir este valor conforme necessário

x = np.arange(len(labels))

fig, ax = plt.subplots(figsize=(10, 6))
line1, = ax.plot(x, values_tree, label='Árvore de Decisão', marker='o')
line3, = ax.plot(x, values_linear, label='Regressão Linear', marker='o')

ax.set_ylabel('Valor')
ax.set_title('Métricas de Desempenho por Modelo')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Adicionando rótulos para os pontos de dados
for i, txt in enumerate(values_tree):
    ax.annotate(f'{txt:.2f}', (x[i], values_tree[i]), textcoords="offset points", xytext=(0,10), ha='center')


for i, txt in enumerate(values_linear):
    ax.annotate(f'{txt:.2f}', (x[i], values_linear[i]), textcoords="offset points", xytext=(0,10), ha='center')
```
